We are going to explore the data set using a **user defined function** `explore_data()` that you can repeatedly use to print rows in a readable way.

Using the function `explore_data()` we can perform two tasks :

> Prints the first few rows of the data set.

> Find the number of rows and columns of each data set by setting the parameter or argument `rows_and_columns` (Which is `False` by default)  as `True`.

````
# defining the function explore_data()

def explore_data(dataset, start, end, rows_and_columns=False):
    """
    The explore_data() function:

    Takes in four parameters:
        dataset --> which is expected to be a list of lists.
        
        start and end --> which are both expected to be integers and represent the starting and the ending indices of a slice from the data set.
        
        rows_and_columns --> which is expected to be a Boolean and has False as a default argument.

    """
    dataset_slice = dataset[start:end]    
    for row in dataset_slice:
        print(row)
        print('\n') # adds a new (empty) line after each row

    if rows_and_columns:
        print('Number of rows:', len(dataset))
        print('Number of columns:', len(dataset[0]))
````

### Finding the Index of the Rows with Missing Data
-------------------------------------------------------------------------------------------------------------------
To find the missing value we need to import numpy so that we could figure out the **NaN** value.

The we need to **create** a user defined function called `get_missing_data_index()` which will return the list of index of the rows with NaN/ Missing Value in it.

````
import numpy 

def get_missing_data_index(data_set) :
    """
    The get_missing_data_index() function :
       
       Takes in one parameter;
       
       data_set --> the list of lists form of the CSV file Data Sets.
       
       Returns;
       
       List of index of the rows with NaN/Missing values in it.
    """
    list_missing_index = []
    for i in range(1, len(data_set)) :
        for j in data_set[i] :
            if j == "" or j == numpy.nan :
               list_missing_index.append(i)
    return list_missing_index
````

## Finding and Removing the Duplicate Apps
----------------------------------------------------------------------------------------------------------------------
We may suspect that there may be duplicates with same App name. 

So, we need to explore the two data sets throughly for the duplicates and create two lists.

> One list should contain the list of **unique/original apps** and the varaible is called as `unique_apps`.

> The second list should contain the list of **dulplicate apps** and the varaible is called as `duplicate_apps`.

To implement the above method we define a user defined function called `find_duplicates()`.

````
#Function Definition for find_duplicates()

def find_duplicates(data_set, unique_apps, duplicate_apps) :
    """
    The find_duplicates() function :
    
       Input :
       It takes in three parameters; 
         
         data_set --> the list of lists form of the CSV file Data Sets.
         
         unique_apps --> which is a empty list and would be updated with unique/original apps names after the execution of the function.
         
         duplicate_apps --> which is a empty list and would be updated with duplicate apps names after the execution of the function.
    
       Output :
       
       Prints the total number of duplicate apps in the given data set, prints few duplicate apps for providing example and
       it updates the variables(empty list) provided to unique_apps and duplicate_apps
    """
    for i in range(1, len(data_set)) :
        if data_set[i][0] not in unique_apps :
            unique_apps.append(data_set[i][0])
        else :
            duplicate_apps.append(data_set[i][0])
            
    print("The total number of duplicate apps is", len(duplicate_apps), "\n")
    print("Few examples of the duplicate apps names are \t\t\t\t\t\t\t\t\t\t", duplicate_apps[:15])
````

So, now to create a new list of lists containg only **unique** app names and with higher number of **reviews** we need to create a new user defined function called `unique_app_name_list()` which will **update** the dictionary where the `key` would be the name of the app and the `value` would be the maximum number of reviews for that app.

````
#Function definition for the function unique_app_name_list()

def unique_app_name_list(data_set, max_review) :
    """
    The unique_app_name_list() function :
    
       Input :
       It takes in two parameters; 
          
          data_set --> the list of lists form of the CSV file Data Sets.
          
          unique_apps --> which is a empty dictionary and would be updated with unique/original apps names as keys and maximum reviews as values after the execution of the function.
       
       Output :
          
          Ouput None is printed or returned
          It just Updates the max_review dictionary 
    """
    for i in range(1, len(data_set)) :
        if data_set[i][0] not in max_review :
            max_review[data_set[i][0]] = data_set[i][3]
        else :
            temp = data_set[i][3]
            if temp > max_review[data_set[i][0]] :
                max_review[data_set[i][0]] = temp
````

Now we are going to implement the above defined function for the `googleplaystore.csv` CSV file.

# Removing the Non-English apps Present in the Data Set
----------------------------------------------------------------------------------------------------------------------
Remember we use English for the apps we develop at our company, and we'd like to analyze only the apps that are directed toward an English-speaking audience. However, if we explore the data long enough, we'll find that both data sets have apps with names that suggest they are not directed toward an English-speaking audience.

So we need remove the Non-English apps to yield better and accurate analysis.

We're not interested in keeping these apps, so we'll remove them. One way to go about this is to remove each app with a name containing a symbol that is not commonly used in English text — English text usually includes letters from the English alphabet, numbers composed of digits from 0 to 9, punctuation marks (., !, ?, ;), and other symbols (+, *, /).

All these characters that are specific to English texts are encoded using the ASCII standard. Each ASCII character has a corresponding number between 0 and 127 associated with it, and we can take advantage of that to build a function that checks an app name and tells us whether it contains non-ASCII characters.

We built this function below, and we need to use `alphabet-detector` python library present in `Github` to find the major language used in the function.

**About** `alphabet-detector` **python library :**

>A library to detect what alphabet something is written in. Works on Python 2.7+ and 3.3+

**Installation** :

>pip install alphabet-detector

**Usage** :

>To instantiate an AlphabetDetector (the object is used for speed optimization):
```
     from alphabet_detector import AlphabetDetector
     ad = AlphabetDetector()
```

We are trying to use the functions defined `alphabet-detector` python library to detect the major language used in the app because some english apps may conttain some emojis or other special characters which may not be in the 0-126 ASCII range. So if we consider the characters present in between ASCII 0 to 126 then some of the Native english apps containg emojis and special characters in them may be left behind and this will have negative effect on our analysis.

To learn more about `alphabet-detector` vist this [link](https://github.com/EliFinkelshteyn/alphabet-detector)

Now we need to create a fucnction which returns `True` if the app is an english app and it should return `False` if the app is a Non_English app.

Let’s start defining the function...

````
#first we need to import TextBlob module from textblob
from alphabet_detector import AlphabetDetector
ad = AlphabetDetector()


#then define the function
def is_english(text) : #text is a variable which has the name of the app in it and it has to be tested
    if ad.only_alphabet_chars(text , "LATIN") :
        return True #True is returned if name of the app is in English(Predominately)
    else :
        return False #Else False is returned
````

As we have designed the `is_english()` function, now we need to check whether the function give correct output for various inputs.

````
#checking for an English app
print("Is 'Instagram' a English app : ", is_english("Instagram"),"\n")

#checking for an Non-English app
print("Is '爱奇艺PPS -《欢乐颂2》电视剧热播' a English app : ", is_english("爱奇艺PPS -《欢乐颂2》电视剧热播"),"\n") 

#checking for an English app with special character like ™
print("Is 'Docs To Go™ Free Office Suite' a English app : ", is_english("Docs To Go™ Free Office Suite"),"\n") 

#checking for an English app with emoji
print("Is 'Instachat 😜' a English app: ", is_english("Instachat 😜"),"\n") 
````

We have successfully defined a function called `is_english()` which return `True` if the app name is in english and `False` otherwise.

Now, our task is to **filter out/ separate** the English apps from the Non-English apps.

For this task we need to use the previously created `is_english()` function to filter out non-English apps from both data sets. Loop through each data set. If an app name is identified as English, append the whole row to a separate list.

To achive this we need to create two empty lists;
 >One containing the list English apps only.
 
 >The other one, containing the list of Non-English apps.

````
#We are going to implement the above mentioned task for the "googleplaystore.csv" Data Set

#Defining two emplty lists
android_clean_english_app = [] #Variable to store list of English apps

android_clean_non_english_app = [] #Variable to store list of Non-English apps

for i in range(len(android_clean)) :
    app_name = android_clean[i][0]
    #print(app_name)
    if is_english(app_name) :
        android_clean_english_app.append(android_clean[i])
    else :
        android_clean_non_english_app.append(android_clean[i])

print("The total number of English apps : ", len(android_clean_english_app), "\n")
print("The total number of Non-English apps : ", len(android_clean_non_english_app), "\n")
print("The total number of Apps in 'android_clean' : ", len(android_clean), "\n")

#checking whether the all the apps in android_clean have been segregated properly
if len(android_clean) == len(android_clean_english_app) + len(android_clean_non_english_app) :
    print("True, the apps have been segregated properly")
````

We'll build two functions we can use to analyze the frequency tables:

- One function to generate frequency tables that show percentages
  
- Another function we can use to display the percentages in a descending order

However, dictionaries don't have order, and it will be very difficult to analyze the frequency tables. We'll need to build a second function that can help us display the entries in the frequency table in descending order.

Now we need to create a function named `freq_table()` that takes in two inputs: `dataset` (which will be a list of lists) and `index` (which will be an integer).

> The function should return the frequency table (as a dictionary) for any column we want. The frequencies should also be expressed as percentages.

````
#defining the function freq_table()

def freq_table(dataset, index) :
    
    Dict = {} #Creating a dict to store the total value of the key
    Total = 0 #A variable to store toatl nunmber of records present under the particular column name

    for apps in dataset :
        Total += 1
        if apps[index] not in Dict :
            Dict[apps[index]] = 1
        else :
            Dict[apps[index]] += 1
    
    #Creating a dictionary to use it as a frequency table
    Frequency_table = {}
    for keys in Dict :
        Percentage = (Dict[keys]/Total) * 100
        Frequency_table[keys] = Percentage

    return Frequency_table
````

Using the workaround above, we wrote a helper function named `display_table()`, which will be able to combine with the function which we wrote in the last code cell called `freq_table()` The `display_table()` function does the following :

- Takes in two parameters: `dataset` and `index`. dataset will be a list of lists, and index will be an integer

- Generates a frequency table using the `freq_table()` function.

- Transforms the frequency table into a list of tuples, then sorts the list in a descending order

- Prints the entries of the frequency table in descending order

````
#defining the function display_table()

def display_table(dataset, index):
    table = freq_table(dataset, index)
    table_display = []
    for key in table:
        key_val_as_tuple = (table[key], key)
        table_display.append(key_val_as_tuple)

    table_sorted = sorted(table_display, reverse = True)
    for entry in table_sorted:
        print(entry[1], ':', entry[0])
````

Now loop over the unique genres of the App Store dataset. For each iteration (below, we'll assume that the iteration variable is named genre):

- Initiate a variable named `total` with a value of 0. This variable will store the sum of user ratings (the number of ratings, not the actual ratings) specific to each genre.
  
- Initiate a variable named `len_genre` with a value of 0. This variable will store the number of apps specific to each genre.
  
- Loop over the App Store dataset, and for each iteration, do the following:
    - Save the app genre to a variable named genre_app.
  
    - If `genre_app` is the same as genre (the iteration variable of the main loop), then do the following:
        - Save the number of user ratings of the app as a float.
  
        - Add up the number of user ratings to the total variable.
        - Increment the `len_genre` variable by 1.
- Compute the average number of user ratings by dividing total by `len_genre`. You should do this outside the nested loop.
- Print the app genre and the average number of user ratings. You should also do this outside the nested loop.

````
# Creating a list to store the list of genres and its corresponding avg_reviews
list_genres_ios = [] 
list_genres_corresponding_avg_rev_ios = []
list_genres_corresponding_percentage_share_of_apps_ios = []

for genres in ios_frequency_table : #Iterating over the keys of the frequency table
    
    total = 0 # A variable to store the Avg of Total User Reviews for that Particular Genre in App Store
    len_genre = 0 #A variable to store number of apps specific to each genre

    for app in ios_final_free :

        index = ios_list[0].index("prime_genre") # Finding the index for prime_genre
        genre_app = app[index]

        if genre_app == genres :
            Index = ios_list[0].index("rating_count_tot") #Finding the index for rating_count_tot
            n_user_rating = float(app[Index])
            total += n_user_rating
            len_genre += 1

    avg_n_user_rating = total / len_genre
    print(genres ,":", avg_n_user_rating, )
    list_genres_ios.append(genres)
    list_genres_corresponding_avg_rev_ios.append(avg_n_user_rating)
    list_genres_corresponding_percentage_share_of_apps_ios.append(ios_frequency_table[genres])
````
